## Definicion de Librerias, (no para calculo de multiplicacion de matrices)


```python
import random # Importamos la biblioteca random para generar n√∫meros aleatorios
import time # Necesario para medir tiempos de ejecuci√≥n
```

# Inciamos la Definicion de Funciones

#### Funci√≥n para Generar Matrices Aleatorias

Esta funci√≥n se encarga de crear una matriz cuadrada ($n \times n$) con elementos generados aleatoriamente.

**Caracter√≠sticas:**

* Crea una matriz cuadrada de dimensiones $n \times n$.
* Los elementos de la matriz se inicializan con valores enteros aleatorios.
* Dichos valores aleatorios est√°n comprendidos entre `valor_min` y `valor_max` (ambos inclusive).

```python
def crear_matriz_aleatoria(n, valor_min=0, valor_max=9):
    matriz = []
    for _ in range(n):
        fila_actual = []
        for _ in range(n):
            elemento_aleatorio = random.randint(valor_min, valor_max)
            fila_actual.append(elemento_aleatorio)
        matriz.append(fila_actual)
    return matriz
```

### Funcion para Imprimir la Matriz en pantalla de manera amistosa

```python
def imprimir_matriz(matriz, nombre_matriz="Matriz"):
    print(f"\n{nombre_matriz}:")
    if not matriz or not matriz[0]:
        print("Matriz vac√≠a o inv√°lida.")
        return
    ancho_columna = 4
    for fila in matriz:
        for elemento in fila:
            print(str(elemento).rjust(ancho_columna), end=" ")
        print()
```

### Definimos la Funci√≥n para el C√°lculo de Matrices

* Multiplica dos matrices cuadradas (`matriz_a` * `matriz_b`).
* Las matrices deben ser de la misma dimensi√≥n $n \times n$.
* No utilizamos ninguna biblioteca que automatice la multiplicaci√≥n.

---

Posteriormente, se realiza una verificaci√≥n b√°sica de dimensiones (aunque el problema asume que las matrices son cuadradas y de dimensiones correctas).

### **Hemos modificado para poder Devolver la matriz resultado y el conteo de ejecuciones del ciclo m√°s interno.**

```python
def multiplicar_matrices(matriz_a, matriz_b):
    n = len(matriz_a)
    if n == 0 or len(matriz_a[0]) != n or len(matriz_b) != n or len(matriz_b[0]) != n:
        print("Error: Las matrices no son cuadradas o no tienen las mismas dimensiones.")
        return [], 0
    matriz_resultado = []
    for _ in range(n):
        fila_resultado_ceros = [0] * n
        matriz_resultado.append(fila_resultado_ceros)
    contador_ciclo_interno = 0
    for i in range(n):
        for j in range(n):
            suma_para_elemento_ij = 0
            for k in range(n):
                suma_para_elemento_ij += matriz_a[i][k] * matriz_b[k][j]
                contador_ciclo_interno += 1
            matriz_resultado[i][j] = suma_para_elemento_ij
    return matriz_resultado, contador_ciclo_interno
```

# --- Nueva Rutina para Pruebas Autom√°ticas ---

Este bloque de c√≥digo Python define una funci√≥n llamada `ejecutar_prueba_automatica`. Su prop√≥sito es configurar y ejecutar pruebas autom√°ticas, probablemente para analizar el rendimiento de alguna operaci√≥n en funci√≥n de un tama√±o de entrada `n`.

**Puntos Clave:**

* **`valores_n_solicitados`**: Se crea una secuencia de valores para `n` que va desde 10 hasta 10000, increment√°ndose de 10 en 10. Estos ser√°n los diferentes tama√±os de entrada para las pruebas.
* **`max_n_ejecucion_real`**: Se establece un l√≠mite (actualmente en 200) para la ejecuci√≥n "completa" de una operaci√≥n (presumiblemente una multiplicaci√≥n, dado el comentario). Si `n` supera este umbral, se estima que el costo computacional es $n^3$ en lugar de ejecutar la operaci√≥n real, para evitar tiempos de ejecuci√≥n excesivos.
* **`tabla_resultados`**: Se inicializa una lista vac√≠a que, se infiere, almacenar√° los resultados obtenidos de cada prueba ejecutada o calculada.
* **Mensaje de Inicio**: La funci√≥n comienza imprimiendo un mensaje que indica el inicio de la ejecuci√≥n autom√°tica y el conteo de operaciones.

En resumen, esta funci√≥n prepara el entorno para una serie de pruebas donde se var√≠a el tama√±o de `n`. Para valores peque√±os de `n`, se ejecuta una operaci√≥n completa, mientras que para valores grandes se utiliza una estimaci√≥n ($n^3$) para predecir el n√∫mero de operaciones, optimizando el tiempo total de prueba. Los resultados se guardar√°n en `tabla_resultados`.

```python
# --- Nueva Rutina para Pruebas Autom√°ticas ---
def ejecutar_prueba_automatica():
    print("\n--- Ejecuci√≥n Autom√°tica y Conteo de Operaciones ---")

    # Definimos la secuencia de n y el l√≠mite para ejecuci√≥n real
    # n_valores va de 10 en 10 hasta 10000.
    valores_n_solicitados = range(10, 10001, 10)
    # L√≠mite pr√°ctico para ejecutar la multiplicaci√≥n completa.
    # Para n > max_n_ejecucion_real, calcularemos n^3 directamente.
    max_n_ejecucion_real = 200 # Aqui podemos ajustar valores mas grandes de 250 se pone duro

    tabla_resultados = []

    print(f"Se generar√°n datos para n en la secuencia: 10, 20, ..., 10000.")
    print(f"La multiplicaci√≥n completa se ejecutar√° para n <= {max_n_ejecucion_real}.")
    print(f"Para n > {max_n_ejecucion_real}, el conteo de operaciones (n^3) se calcular√° directamente.\n")

    for n_actual in valores_n_solicitados:
        print(f"Procesando para n = {n_actual}:")
        conteo_operaciones = 0
        tiempo_ejecucion_str = ""

        if n_actual <= max_n_ejecucion_real:
            # Crear matrices peque√±as (contenido no es cr√≠tico para el conteo)
            matriz_a_dummy = crear_matriz_aleatoria(n_actual, 0, 1)
            matriz_b_dummy = crear_matriz_aleatoria(n_actual, 0, 1)

            tiempo_inicio = time.time()
            # Ejecutamos la multiplicaci√≥n para obtener el conteo real
            _, conteo_operaciones = multiplicar_matrices(matriz_a_dummy, matriz_b_dummy)
            tiempo_fin = time.time()

            tiempo_transcurrido = tiempo_fin - tiempo_inicio
            tiempo_ejecucion_str = f"{tiempo_transcurrido:.4f}"
            print(f"  -> Ejecuci√≥n real: Conteo={conteo_operaciones}, Tiempo={tiempo_ejecucion_str}s")
        else:
            # Para n grandes, calculamos n^3 directamente
            conteo_operaciones = n_actual**3
            tiempo_ejecucion_str = "N/A (calculado)"
            print(f"  -> C√°lculo directo (n^3): Conteo={conteo_operaciones}")

        tabla_resultados.append({
            "n": n_actual,
            "conteo": conteo_operaciones,
            "tiempo_s": tiempo_ejecucion_str
        })

    # Imprimir la tabla de resultados
    print("\n\n--- Tabla de Resultados (n vs. Conteo de Operaciones del Ciclo Interno) ---")
    print("------------------------------------------------------------------------------")
    # Ajustar anchos de columna seg√∫n sea necesario
    print(f"| {'n':^5} | {'Conteo Operaciones':^30} | {'Tiempo (s)':^20} |")
    print("------------------------------------------------------------------------------")
    for resultado in tabla_resultados:
        n_str = str(resultado['n']).rjust(5)
        conteo_str = str(resultado['conteo']).rjust(30)
        tiempo_str = str(resultado['tiempo_s']).rjust(20)
        print(f"| {n_str} | {conteo_str} | {tiempo_str} |")
    print("------------------------------------------------------------------------------")
```

### Funci√≥n Principal

Esta es la Funci√≥n Principal que nos va a ayudar a tener un Men√∫ principal e invocar a nuestras funciones que hemos definido para:

* Creaci√≥n aleatoria de Matriz
* Impresi√≥n de Matriz
* Multiplicaci√≥n de matriz

> Funci√≥n principal que maneja el men√∫ y el flujo del programa.


### En `main()`:

* La llamada a `multiplicar_matrices` ahora desempaqueta dos valores: `matriz_producto, conteo_operaciones_interno = multiplicar_matrices(matriz_1, matriz_2)`.
* Despu√©s de imprimir la matriz resultado, se a√±ade una l√≠nea para mostrar el valor de `conteo_operaciones_interno`.
* Tambi√©n se a√±ade una peque√±a nota que indica que este conteo es $n^3$, lo cual es esperado para este algoritmo.

# ******CAMBIOS A CONSIDERAR DE ESTA VERSION******

## Modificaci√≥n en `main()`:
Se ha reestructurado el men√∫ para ofrecer tres opciones:
* Multiplicaci√≥n manual (como antes).
* La nueva rutina de prueba autom√°tica.
* Salir.

El c√≥digo de la multiplicaci√≥n manual se ha movido bajo la `opcion == '1'`.

```python
# --- Programa Principal (Menusito Modificado) ---
def main():
    print("=====================================================")
    print("  Multiplicador de Matrices Cuadradas (n x n)  ")
    print("=====================================================")

    while True:
        print("\n-------------------- MEN√ö -----------------------")
        print("1. Introducir n y multiplicar matrices manualmente")
        print("2. Ejecutar prueba autom√°tica (n vs conteo de operaciones)")
        print("3. Salir")
        opcion = input("Selecciona una opci√≥n: ")

        if opcion == '1':
            entrada_usuario = input("Introduce el tama√±o (n) para las matrices cuadradas: ")
            try:
                n = int(entrada_usuario)
                if n <= 0:
                    print("El valor de n debe ser un entero positivo.")
                    continue
                # (Opcional) Advertencia para n grande en modo manual
                if n > 50: # Un l√≠mite m√°s alto para modo manual que para autom√°tico masivo
                    print(f"Advertencia: n={n} es grande, la generaci√≥n e impresi√≥n puede ser extensa.")
                    confirmacion = input("¬øDeseas continuar? (s/N): ")
                    if confirmacion.lower() != 's':
                        continue

                print(f"\nGenerando dos matrices ({n}x{n}) con elementos aleatorios (0-9)...")
                matriz_1 = crear_matriz_aleatoria(n, 0, 9)
                matriz_2 = crear_matriz_aleatoria(n, 0, 9)

                imprimir_matriz(matriz_1, "Matriz A")
                imprimir_matriz(matriz_2, "Matriz B")

                print("\nCalculando el producto (Matriz A * Matriz B)...")
                matriz_producto, conteo_ops = multiplicar_matrices(matriz_1, matriz_2)

                if matriz_producto:
                    imprimir_matriz(matriz_producto, "Matriz Resultado (A * B)")
                    print(f"\nEl ciclo m√°s interno (k) se ejecut√≥ {conteo_ops} veces.")
                    if n > 0:
                        print(f"(Esto es igual a n*n*n = {n}*{n}*{n} = {n**3})")

            except ValueError:
                print("Entrada inv√°lida. Debes introducir un n√∫mero entero para n.")
                continue

        elif opcion == '2':
            ejecutar_prueba_automatica()

        elif opcion == '3':
            print("¬°Gracias por usar el programa! Adi√≥s.")
            break

        else:
            print("Opci√≥n no v√°lida. Por favor, intenta de nuevo.")

if __name__ == "__main__":
    main()
```
