Crea un programa que pida al usuario que ingrese su ingreso anual y calcule el impuesto que debe pagar seg√∫n las siguientes tasas:

* Menos de 10,000: 0%
* De 10,000 a 20,000: 10%
* De 20,001 a 50,000: 20%
* M√°s de 50,000: 30%

**Instrucciones:**

* Pide al usuario que ingrese su ingreso anual.
* Usa `if` para verificar si el ingreso es un n√∫mero negativo.
* Usa `if-elif-else` para determinar el porcentaje de impuesto y calcular el monto a pagar.
* Recuerde enviar su `.py` y capturas de pantalla de los resultados obtenidos

---

```python
def calcular_impuesto(ingreso_anual):
    """Funcion para hacer el Calculo de las tasas"""

    #aqui inicializo las variables para que no se contaminen si las vuelvo a usar
    tasa_impuesto = 0
    impuesto_a_pagar = 0


    #Usando IF/ELIF/ELSE hacemos muy facil la determinacion de la tasa de impuesto
    if ingreso_anual < 0:
        print("Error: El ingreso anual no puede ser un n√∫mero negativo.")
        tasa_impuesto = 0
        impuesto_a_pagar = 0
    elif ingreso_anual < 10000:
        tasa_impuesto = 0.00
    elif ingreso_anual <= 20000:
        tasa_impuesto = 0.10
    elif ingreso_anual <= 50000:
        tasa_impuesto = 0.20
    else:  # M√°s de 50,000
        tasa_impuesto = 0.30


    # Solo calculamos si el monto es Positivo, si es negativo, pues nada
    if ingreso_anual >= 0:
        impuesto_a_pagar = ingreso_anual * tasa_impuesto

    #regresamos de la funcion los valores de la tasa que le corresponde y el varo a pagar
    return tasa_impuesto, impuesto_a_pagar

def main():
    """Creamos esta funcion principal para poder hacer muchos calculos"""
    print("Programa de C√°lculo de Impuestos Anuales")
    print("----------------------------------------")

    for i in range(10): #voy hacer 10 calculos
        print(f"\n--- C√°lculo de Impuestos #{i + 1} ---")

        try:
            ingreso_str = input(f"({i + 1}/5) Indique $$$$$, ingrese su ingreso anual: $")
            ingreso_anual_usuario = float(ingreso_str)

            tasa, impuesto = calcular_impuesto(ingreso_anual_usuario)

            print(f"Ingreso Anual Ingresado: ${ingreso_anual_usuario:,.2f}")
            if ingreso_anual_usuario >= 0: # Solo mostrar tasa e impuesto si el ingreso es v√°lido
                print(f"Tasa de Impuesto Aplicada: {tasa*100:.0f}%")
                print(f"Impuesto a Pagar: ${impuesto:,.2f}")
            # El mensaje de error para ingresos negativos ya se maneja en calcular_impuesto

        except ValueError:
            print("Error: Valor Invalido. mete tu salario en un monto real porfa.")
        except Exception as e:
            print(f"Ha ocurrido un error inesperado FATALITYYYY: {e}")

        print("----------------------------------------")

    print("\nSe han completado los 10 c√°lculos de impuestos.")

    print("\nA tope con el propedeutico.")

if __name__ == "__main__":
    main()
```
